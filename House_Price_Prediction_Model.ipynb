#### Checking and removing the outliers of Price column

```python
sns.boxplot(df['price'])
plt.title('Price')
plt.show()
```

```python
## Price column has outliers and it's skewed so we'll use IQR method to handle outliers

Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1
IQR
```

```python
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR
lower_limit,upper_limit
```

```python
def limit(value):
    if value<lower_limit:
        return lower_limit
    if value>upper_limit:
        return upper_limit
    else:
        return value
```

```python
df['price']=df['price'].apply(limit)
```

```python
sns.boxplot(df['price']) 
plt.title('Price after removing the outliers')
plt.show()
```

#### Encoding zipcode and yr_built columns into ordinal categorical groups based on average prices .

```python
df['zipcode'].value_counts().shape
```

```python
## calculating average sale price across all zipcode .

zip_table = df.groupby('zipcode').agg({'price':'mean'}).sort_values('price',ascending=True)
zip_table
```

```python
## Create a new column 'zipcode_group' that bins the average prices into 10 groups with custom labels.

zip_table['zipcode_group'] = pd.cut(zip_table['price'],bins=10,
                   labels=['Zipcode_Group_0',
                           'Zipcode_Group_1',
                           'Zipcode_Group_2',
                           'Zipcode_Group_3',
                           'Zipcode_Group_4',
                           'Zipcode_Group_5',
                           'Zipcode_Group_6',
                           'Zipcode_Group_7',
                           'Zipcode_Group_8',
                           'Zipcode_Group_9',
                           ],include_lowest=True)
```

```python
zip_table.head()
```

```python
zip_table['zipcode_group'].value_counts()
```

```python
## Merging dataframes.

df = pd.merge(df,zip_table,left_on='zipcode',how='left',right_index=True)
df.head()
```

```python
## One-hot encoding the 'zipcode_group' column into dummy variables, dropping the first category to avoid multicollinearity.

df = pd.get_dummies(df,columns = ['zipcode_group'],drop_first=True,dtype='int')
```

```python
df.head(2)
```

```python
## calculating average sale price across all yeard of built .

yr_built_table = df.groupby('yr_built').agg({'price_x':'mean'}).sort_values('price_x',ascending=True)
yr_built_table
```

```python
## Create a new column 'yr_built_group' that bins the average prices into 6 groups with custom labels.

yr_built_table['yr_built'] = pd.qcut(yr_built_table['price_x'],q=6,
                   labels=['year_Group_0',
                           'year_Group_1',
                           'year_Group_2',
                           'year_Group_3',
                           'year_Group_4',
                           'year_Group_5'])
```

```python
yr_built_table['yr_built'].value_counts()
```

```python
yr_built_table.sample(10)
```

```python
## Merging the dataframes

df = pd.merge(df,yr_built_table,left_on='yr_built',how='left',right_index=True)
df.head()
```

```python
## One-hot encoding the 'yr_built_y' column into dummy variables, dropping the first category to avoid multicollinearity.

df = pd.get_dummies(df,columns = ['yr_built_y'],drop_first=True,dtype='int')
df.head()
```

```python
df.rename(columns={'price_x_x':'price'},inplace=True)
```

```python
df.columns
```

```python
## Dropping irrelevant columns

df=df.drop(columns=['zipcode','price_x_y','price_y','yr_built','yr_built_x','date'])
```

```python
df.columns
```

```python
## Creating dependent and independent variables .

X = df.drop(columns=['price'])
y = df['price']
```

```python
X.head()
```

```python
y.head()
```

#### Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=22)
```

```python
X_train.shape,X_test.shape
```

```python
y_train.shape,y_test.shape
```

```python
X_train.head()
```

#### Training the model

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(n_estimators=200)
```

```python
rf.fit(X_train,y_train)
```

```python
y_pred = rf.predict(X_test)
```

```python
from sklearn.metrics import r2_score , root_mean_squared_error , mean_absolute_error
```

```python
print(f'R2 Score: {round(r2_score(y_test,y_pred),2)}')
print(f'Root Mean Squared Error: {round(root_mean_squared_error(y_test,y_pred),2)}')
print(f'Mean Absolute Error: {round(mean_absolute_error(y_test,y_pred),2)}')
```

#### Extracting the feature importances from the fitted Random Forest model to analyze the contribution of each feature.


```python
importances = rf.feature_importances_
imp_table = pd.DataFrame({"columns":X_train.columns,"Importances":importances})
imp_table
```

```python
plt.barh(imp_table['columns'],imp_table['Importances'])
plt.xlabel('Importances')
plt.ylabel('Features')
plt.title('Feature Importance in Random Forest')
plt.show()
```
